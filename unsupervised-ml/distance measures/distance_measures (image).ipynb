# Distance Measures on Image

Steps :
1. Resize the image
2. Convert the matrix to 1D array
3. Calculate distance measures

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("img1.jpeg")
img2 = cv2.imread("img2.jpeg")
```

```python
# Convert BGR to RGB (for display)
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img1_rgb)
plt.axis("off")
plt.title("Image 1")

plt.subplot(1,2,2)
plt.imshow(img2_rgb)
plt.axis("off")
plt.title("Image 2")

plt.show()
```

```python
# Print sizes
print("Size of image 1 :", img1.shape)
print("Size of image 2 :", img2.shape)
```

```python
# Convert to float and normalize
v1 = img1.astype(np.float32).flatten() / 255.0
v2 = img2.astype(np.float32).flatten() / 255.0
```

```python
v1.shape
```

```python
v2.shape
```

Distance Measures

```python
# Euclidean Distance
euclidean = np.linalg.norm(v1 - v2)
print("Euclidean Distance:", euclidean)

# Manhattan Distance
manhattan = np.sum(np.abs(v1 - v2))
print("Manhattan Distance:", manhattan)

# Minkowski Distance (p = 3)
p = 3
minkowski = np.power(np.sum(np.abs(v1 - v2) ** p), 1/p)
print(f"Minkowski Distance (p={p}):", minkowski)
```

```python
mse = np.mean((v1 - v2) ** 2)
print("MSE:", mse)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

v1_cs = v1.reshape(1, -1)
v2_cs = v2.reshape(1, -1)

cos_sim = cosine_similarity(v1_cs, v2_cs)[0][0]
print("Cosine Similarity:", cos_sim)
```

```python
chebyshev = np.max(np.abs(v1 - v2))
print("Chebyshev Distance:", chebyshev)
```

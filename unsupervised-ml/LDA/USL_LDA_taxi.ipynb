## Linear Discriminant Analysis (LDA)

- a supervised machine learning and statistical technique used primarily for classification and dimensionality reduction

- it projects high dimensional data onto a lower dimensional space while maximizing the class separability. (2D or more to 1D)

- criteria used by LDA to create a new axis :
  - **maximizes** the distance between means of two classes and
  - **minimizes** the variation within each class


# With Library

```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
```

```python
# Load taxi dataset
taxi_df = sns.load_dataset('taxis')
taxi_df.head()
```

```python
taxi_df.info()
```

```python
taxi_df.describe()
```

```python
# check for missing values
taxi_df.isnull().sum()
```

```python
# Drop missing values
taxi_df = taxi_df.dropna()
```

```python
taxi_df.isnull().sum()
```

```python
LE = LabelEncoder()
taxi_df['payment'] = LE.fit_transform(taxi_df['payment'])
```

### Feature Selection

```python
# Select features for classification
X = taxi_df[['fare', 'distance', 'pickup_borough', 'dropoff_borough']]
X = pd.get_dummies(X, drop_first=True)

y = taxi_df['payment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=67)
```

```python
SS = StandardScaler()
X_train_scaled = SS.fit_transform(X_train)
X_test_scaled = SS.transform(X_test)
```

### Accuracy with and without LDA

```python
RFC = RandomForestClassifier(n_estimators=100,
                             random_state=69)
RFC.fit(X_train_scaled, y_train)
```

```python
y_pred = RFC.predict(X_test_scaled)
print("Accuracy on raw data =", metrics.accuracy_score(y_test, y_pred))
```

```python
LDA = LinearDiscriminantAnalysis(n_components=1)
X_train_lda = LDA.fit_transform(X_train_scaled, y_train)
X_test_lda = LDA.transform(X_test_scaled)

RFC.fit(X_train_lda, y_train)

y_pred_lda = RFC.predict(X_test_lda)
print("Accuracy using LDA (1 component) =", metrics.accuracy_score(y_test, y_pred_lda))
```

```python
print(f"Accuracy on raw data = {metrics.accuracy_score(y_test, y_pred):.4f}")
print(f"Accuracy using LDA (1 component) = {metrics.accuracy_score(y_test, y_pred_lda):.4f}")
```

The **decrease in accuracy** indicates that dimensionality reduction led to **information loss**.

Since the Taxi dataset may not satisfy LDA assumptions such as linear separability and equal covariance, reducing the data to one dimension reduced the discriminative power of the model.

Therefore, LDA did not improve classification performance.

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix WITHOUT LDA")
print(confusion_matrix(y_test, y_pred))

print("Confusion Matrix WITH LDA")
print(confusion_matrix(y_test, y_pred_lda))
```

# Without Library (Manual Implementation)

### Steps :
1. Calcualte the mean vectors for each class
2. Compute the covariance of each class
3. Compute the Within-Class Scatter Matrix (SW)
4. Compute the Between-Class Scatter Matrix (SB)
5. Find eigen values and choose the top eigen value
6. Find eigen vectors
7. Find the dot prodect of data points with eigen vectors

Assumption:

X → feature matrix (numeric only)

y → binary class (0 and 1)

```python
taxi_df['fare_class'] = (taxi_df['total'] > taxi_df['total'].median()).astype(int)
```

0 → Low fare

1 → High fare

```python
numeric_cols = ['passengers', 'distance', 'fare', 'tip', 'tolls', 'total']

X = taxi_df[numeric_cols].values
y = taxi_df['fare_class'].values
```

```python
print(X.shape)
print(X.dtype)
```

```python
numeric_cols = ['passengers', 'distance', 'fare', 'tip', 'tolls']
```

```python
X = taxi_df[numeric_cols].values
y = taxi_df['fare_class'].values
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Separate Classes
X0 = X[y == 0]
X1 = X[y == 1]
```

```python
# Compute Mean Vectors
mu0 = np.mean(X0, axis=0)
mu1 = np.mean(X1, axis=0)
```

@ operator means matrix multiplication.

```python
# Compute Within-Class Scatter Matrix
S0 = np.zeros((X.shape[1], X.shape[1]))
for x in X0:
    diff = (x - mu0).reshape(-1,1)
    S0 += diff @ diff.T

S1 = np.zeros((X.shape[1], X.shape[1]))
for x in X1:
    diff = (x - mu1).reshape(-1,1)
    S1 += diff @ diff.T

SW = S0 + S1
SW
```

```python
# Compute Between-Class Scatter Matrix
mean_diff = (mu1 - mu0).reshape(-1,1)
SB = mean_diff @ mean_diff.T
SB
```

```python
# Solve Eigenvalue Problem
SW_inv = np.linalg.pinv(SW)

matrix = SW_inv @ SB

eigenvalues, eigenvectors = np.linalg.eig(matrix)
```

```python
# Select Top Eigenvector
idx = np.argsort(abs(eigenvalues))[::-1]

W = eigenvectors[:, idx[0]]
W
```

```python
# Project Data
X_lda = X @ W
X_lda
```

```python
# Classification After Projection
m0 = W @ mu0
m1 = W @ mu1

threshold = (m0 + m1) / 2

y_pred = (X_lda > threshold).astype(int)
```

```python
# Accuracy
accuracy = np.mean(y_pred == y)
print("Manual LDA Accuracy:", accuracy)
```
